```python
import string
from nltk import FreqDist

# load text and convert to lowercase and remove punctuation, and split(similar to tokenize)
input_string = open('C:\\Users\\bless\\PycharmProjects\\practicum\\emma.txt', 'r', encoding='utf-8')
input_string = input_string.read()
input_string = input_string.lower()
input_string = input_string.translate(str.maketrans('','',string.punctuation))


# define a function to get word frequencies
def get_word_freq(input_string):
    input_string_list= input_string.split()
#     print(input_string_list)

    di = []

    for i in input_string_list:
      
        if i not in di:
            di.append(i)
        
#     print(di)
    
    for i in range(0, len(di)):
        count = input_string_list.count(di[i])
        print(di[i], count)

   


def Driver():
    get_word_freq(input_string)
    
if __name__ == '__main__':
    Driver()

    
```

```python
print(input_string_list[0:100])
```

```python
fd = FreqDist()
for word in input_string_list:
    fd[word] += 1
```

```python
most_common_words = map(lambda p: str(p[0]) + " "+ str(p[1]), fd.most_common(20))
for cw in most_common_words:
    print(cw)
```

```python
import pandas as pd

row_list = []
rank = 0

for x in fd.most_common():
    word = x[0]
    count = x[1]
    rank += 1
    
    row = {'Word': word, 'Rank': rank, 'Frequency': count}
    row_list.append(row)
print(row_list[:5])
```

```python
df = pd.DataFrame(row_list, columns = ['Word', 'Rank', 'Frequency'])

print(df.head(20))
```

```python
import random
import string
```

```python
#Define a function to generate random text
def string_gen(str_size, allowed_chars):
    return ''.join(random.choice(allowed_chars) for x in range(str_size))

chars = ('abcdef ')
text = string_gen(100000, chars)
print(text)
```

```python
#tokenize text
from nltk import word_tokenize, sent_tokenize
text = word_tokenize(str(text))

print(text)
```

```python
# determine frequency of words
fd = FreqDist()
for string in text:
    fd[string] += 1
```

```python
most_common_strings = map(lambda p: str(p[0]) + " "+ str(p[1]), fd.most_common(20))
for cw in most_common_strings:
    print(cw)
```

```python
# create a list/table with string, rank and frequency
row_list = []
rank = 0

for x in fd.most_common():
    string = x[0]
    count = x[1]
    rank += 1
    
    row = {'String': string, 'Rank': rank, 'Frequency': count}
    row_list.append(row)
print(row_list[:5])
```

```python
# create a new table/dataframe
df_strings = pd.DataFrame(row_list, columns = ['String', 'Rank', 'Frequency'])

print(df_strings.head(20))
```

```python
# store dataframe as a csv file
df_strings.to_csv(os.path.join(current_path, 'Strings-zipflaw.csv'), index = False, encoding = 'utf-8')

df = None

strings_data = pd.read_csv(os.path.join(current_path, 'Strings-zipflaw.csv'), encoding = 'utf-8')

print(strings_data.head(20))
```
